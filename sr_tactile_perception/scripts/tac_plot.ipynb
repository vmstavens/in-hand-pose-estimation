```python

import numpy as np

window_size   = 10
window        = np.ones(window_size) / window_size
input_labels  = ['f_px', 'f_py', 'f_pz', 'f_f1x', 'f_f1y', 'f_f1z', 'f_f2x', 'f_f2y', 'f_f2z', 'f_f3x', 'f_f3y', 'f_f3z']
output_labels = ["f_pac0","f_pac1","f_pdc","f_tac","f_tdc"]
output_labels.extend([f"f_e{i}" for i in range(1,20)])

path          = "/home/user/projects/shadow_robot/base/src/biotac_sim_plugin/biotac_sim_lib/data/"

def graph_data(df_in, df_out, axes):
	for i, l in enumerate(input_labels):
		df_in[l] = df_in[i].rolling(window_size, center=True).apply(lambda x: np.dot(x, window))
		axes[0].plot(df_in[l], label=l)
	axes[0].legend(ncol=2)

	for i, l in enumerate(output_labels):
		df_out[l] = df_out[i].rolling(window_size, center=True).apply(lambda x: np.dot(x, window))
		axes[1].plot(df_out[l], label=l)
	axes[1].legend(ncol=3)
	return df_in, df_out

# controlled relation

# df_sim_cube_input = pd.read_csv(path + "sim_cube_input.csv", header=None)
# df_sim_cube_output = pd.read_csv(path + "sim_cube_output.csv", header=None)
# fig_cube, axes_cube = plt.subplots(ncols=2, figsize=(10, 5))
# df_in, df_out = graph_data(df_sim_cube_input,df_sim_cube_output,axes_cube)

# df_sim_edge_input = pd.read_csv(path + "sim_edge_input.csv", header=None)
# df_sim_edge_output = pd.read_csv(path + "sim_edge_output.csv", header=None)
# fig_edge, axes_edge = plt.subplots(ncols=2, figsize=(10, 5))
# df_in, df_out = graph_data(df_sim_edge_input, df_sim_edge_output, axes_edge)

# df_sim_sphere_input = pd.read_csv(path + "sim_sphere_input.csv", header=None)
# df_sim_sphere_output = pd.read_csv(path + "sim_sphere_output.csv", header=None)
# fig_sphere, axes_sphere = plt.subplots(ncols=2, figsize=(10, 5))
# df_in, df_out = graph_data(df_sim_sphere_input, df_sim_sphere_output, axes_sphere)

df_sim_corners_input = pd.read_csv(path + "sim_corners_input.csv", header=None)
df_sim_corners_output = pd.read_csv(path + "sim_corners_output.csv", header=None)
fig_corners, axes_corners = plt.subplots(ncols=2, figsize=(10, 5))
df_in, df_out = graph_data(df_sim_corners_input, df_sim_corners_output, axes_corners)

# df_train_input = pd.read_csv(path + "train_input.csv", header=None)
# df_train_output = pd.read_csv(path + "train_output.csv", header=None)
# fig_corners, axes_corners = plt.subplots(ncols=2, figsize=(10, 5))
# df_in, df_out = graph_data(df_train_input, df_train_output, axes_corners)

```
